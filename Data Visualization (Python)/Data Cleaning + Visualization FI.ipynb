```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pandas as pd

df = pd.read_csv(r"C:\Users\mahla\Downloads\DataSets\Superstore.csv", encoding='latin-1')
df
```

```python
print("First 5 rows")
print(df.head())
```

```python
print("\nData Info")
print(df.info())
```

```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nDescriptive Statistics")
print(df.describe())
```

```python
print("\nOUR COLUMNS")
print(df.columns.tolist())
```

```python
# Convert Order Date and Ship Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])  
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Verify the conversion worked
print(df[['Order Date', 'Ship Date']].dtypes)
print("\nDate range:")
print(f"From: {df['Order Date'].min()}")
print(f"To: {df['Order Date'].max()}")
```

```python
#time features for analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter

# Create a Year-Month column for grouping
df['Year_Month'] = df['Order Date'].dt.to_period('M')

# Check what we added
print("New columns added:")
print(df[['Year', 'Month', 'Quarter', 'Year_Month']].head())
print("\nUnique years:", df['Year'].unique())
```

```python
# Calculate profit margin (profit as percentage of sales)
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100

# Flag loss-making transactions
df['Is_Loss'] = df['Profit'] < 0

# Check the results
print("Profit Margin stats:")
print(f"Min: {df['Profit_Margin'].min():.2f}%")
print(f"Max: {df['Profit_Margin'].max():.2f}%")
print(f"Average: {df['Profit_Margin'].mean():.2f}%")
print(f"\nLoss-making transactions: {df['Is_Loss'].sum()} out of {len(df)}")
print(f"Percentage of losses: {(df['Is_Loss'].sum()/len(df))*100:.1f}%")
```

```python
# Calculate key metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_quantity = df['Quantity'].sum()
avg_profit_margin = df['Profit_Margin'].mean()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()

# Create summary
print("="*50)
print("EXECUTIVE SUMMARY")
print("="*50)
print(f"Total Sales:      ${total_sales:,.2f}")
print(f"Total Profit:     ${total_profit:,.2f}")
print(f"Total Quantity:   {total_quantity:,} units")
print(f"Avg Profit Margin: {avg_profit_margin:.1f}%")
print(f"Total Orders:     {total_orders:,}")
print(f"Unique Customers: {total_customers:,}")
print(f"Avg Order Value:  ${total_sales/total_orders:,.2f}")
print(f"Avg Profit/Order: ${total_profit/total_orders:,.2f}")
```

```python
# Group data by month
monthly_trend = df.groupby('Year_Month').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Convert Year_Month to string for plotting
monthly_trend['Date_Str'] = monthly_trend['Year_Month'].astype(str)

# Display first few and last few months
print("First 3 months:")
print(monthly_trend.head(3))
print("\nLast 3 months:")
print(monthly_trend.tail(3))

# Quick visual
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(monthly_trend['Date_Str'], monthly_trend['Sales'], marker='o', label='Sales')
plt.plot(monthly_trend['Date_Str'], monthly_trend['Profit'], marker='s', label='Profit')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Sales and Profit Trend')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyze performance by Category
category_performance = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2).sort_values('Sales', ascending=False)

print("CATEGORY PERFORMANCE")
print("="*50)
print(category_performance)

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sales by category
axes[0].bar(category_performance.index, category_performance['Sales'], color='steelblue')
axes[0].set_title('Total Sales by Category')
axes[0].set_ylabel('Sales ($)')
axes[0].tick_params(axis='x', rotation=45)

# Profit by category (color coded)
colors = ['green' if x > 0 else 'red' for x in category_performance['Profit']]
axes[1].bar(category_performance.index, category_performance['Profit'], color=colors)
axes[1].set_title('Total Profit by Category')
axes[1].set_ylabel('Profit ($)')
axes[1].tick_params(axis='x', rotation=45)
axes[1].axhline(y=0, color='black', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
# Analyze Sub-Category performance
subcat_performance = df.groupby('Sub-Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2).sort_values('Profit', ascending=False)

print("SUB-CATEGORY PERFORMANCE (Top 10 by Profit)")
print("="*60)
print(subcat_performance.head(10))

print("\nSUB-CATEGORY PERFORMANCE (Bottom 5 by Profit)")
print("="*60)
print(subcat_performance.tail(5))

# Visualize top 5 and bottom 5
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Top 5 by profit
top5_profit = subcat_performance.head(5)
axes[0].barh(range(len(top5_profit)), top5_profit['Profit'], color='green')
axes[0].set_yticks(range(len(top5_profit)))
axes[0].set_yticklabels(top5_profit.index)
axes[0].set_xlabel('Profit ($)')
axes[0].set_title('Top 5 Most Profitable Sub-Categories')
axes[0].invert_yaxis()

# Bottom 5 by profit
bottom5_profit = subcat_performance.tail(5)
colors_bottom = ['red' if x < 0 else 'green' for x in bottom5_profit['Profit']]
axes[1].barh(range(len(bottom5_profit)), bottom5_profit['Profit'], color=colors_bottom)
axes[1].set_yticks(range(len(bottom5_profit)))
axes[1].set_yticklabels(bottom5_profit.index)
axes[1].set_xlabel('Profit ($)')
axes[1].set_title('Bottom 5 Least Profitable Sub-Categories')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Analyze Region performance
region_performance = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean',
    'Order ID': 'nunique'
}).round(2).sort_values('Sales', ascending=False)

region_performance.columns = ['Total Sales', 'Total Profit', 'Avg Margin %', 'Number of Orders']
region_performance['Profit per Order'] = region_performance['Total Profit'] / region_performance['Number of Orders']

print("REGION PERFORMANCE")
print("="*70)
print(region_performance)

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sales by region
axes[0].bar(region_performance.index, region_performance['Total Sales'], color='steelblue')
axes[0].set_title('Total Sales by Region')
axes[0].set_ylabel('Sales ($)')
axes[0].tick_params(axis='x', rotation=45)

# Profit by region (color coded)
colors = ['green' if x > 0 else 'red' for x in region_performance['Total Profit']]
axes[1].bar(region_performance.index, region_performance['Total Profit'], color=colors)
axes[1].set_title('Total Profit by Region')
axes[1].set_ylabel('Profit ($)')
axes[1].tick_params(axis='x', rotation=45)
axes[1].axhline(y=0, color='black', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()

# Find best and worst cities
city_performance = df.groupby('City').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).round(2).sort_values('Profit', ascending=False)

print("\nTOP 5 CITIES BY PROFIT:")
print(city_performance.head(5))

print("\nBOTTOM 5 CITIES BY PROFIT:")
print(city_performance.tail(5))
```

```python
# Analyze Customer Segment performance
segment_performance = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Profit_Margin': 'mean',
    'Customer ID': 'nunique',
    'Order ID': 'nunique'
}).round(2).sort_values('Sales', ascending=False)

segment_performance.columns = ['Total Sales', 'Total Profit', 'Avg Margin %', 'Unique Customers', 'Orders']
segment_performance['Profit per Customer'] = segment_performance['Total Profit'] / segment_performance['Unique Customers']
segment_performance['Avg Order Value'] = segment_performance['Total Sales'] / segment_performance['Orders']

print("CUSTOMER SEGMENT PERFORMANCE")
print("="*80)
print(segment_performance)

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Sales by segment
axes[0].bar(segment_performance.index, segment_performance['Total Sales'], color='#3498DB')
axes[0].set_title('Total Sales by Segment')
axes[0].set_ylabel('Sales ($)')
axes[0].tick_params(axis='x', rotation=45)

# Profit by segment
colors = ['green' if x > 0 else 'red' for x in segment_performance['Total Profit']]
axes[1].bar(segment_performance.index, segment_performance['Total Profit'], color=colors)
axes[1].set_title('Total Profit by Segment')
axes[1].set_ylabel('Profit ($)')
axes[1].axhline(y=0, color='black', linestyle='-', linewidth=0.5)
axes[1].tick_params(axis='x', rotation=45)

# Profit per customer
axes[2].bar(segment_performance.index, segment_performance['Profit per Customer'], color='#9B59B6')
axes[2].set_title('Profit per Customer')
axes[2].set_ylabel('Profit per Customer ($)')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Check loss concentration by segment
loss_by_segment = df[df['Is_Loss']].groupby('Segment').agg({
    'Profit': 'sum',
    'Order ID': 'count'
}).round(2)
loss_by_segment.columns = ['Total Loss', 'Number of Loss Transactions']
loss_by_segment['Avg Loss per Transaction'] = loss_by_segment['Total Loss'] / loss_by_segment['Number of Loss Transactions']

print("\nLOSS ANALYSIS BY SEGMENT:")
print("="*50)
print(loss_by_segment)
```

```python
# Analyze loss-making transactions
losses = df[df['Is_Loss']]

# Check discount impact on losses
discount_analysis = losses.groupby('Discount').agg({
    'Profit': 'sum',
    'Sales': 'sum',
    'Order ID': 'count'
}).round(2)

print("IMPACT OF DISCOUNTS ON LOSSES")
print("="*60)
print(discount_analysis.head(10))

# Compare discounts on profitable vs loss-making
print("\nDISCOUNT COMPARISON")
print("="*60)
profitable = df[~df['Is_Loss']]
print(f"Average discount on PROFITABLE transactions: {profitable['Discount'].mean():.2%}")
print(f"Average discount on LOSS-MAKING transactions: {losses['Discount'].mean():.2%}")

# Specific product analysis - Tables
tables_performance = df[df['Sub-Category'] == 'Tables'].agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Discount': 'mean',
    'Order ID': 'count'
}).round(2)

print("\nTABLES PERFORMANCE")
print("="*60)
print(f"Total Sales: ${tables_performance['Sales']:,.2f}")
print(f"Total Profit: ${tables_performance['Profit']:,.2f}")
print(f"Average Discount: {tables_performance['Discount']:.2%}")
print(f"Number of Orders: {tables_performance['Order ID']}")

# Check if high discounts cause the loss
high_discount_tables = df[(df['Sub-Category'] == 'Tables') & (df['Discount'] > 0.3)]
print(f"\nTables sold with >30% discount: {len(high_discount_tables)} orders")
print(f"Profit from these: ${high_discount_tables['Profit'].sum():,.2f}")

# Visualize discount vs profit for Tables
plt.figure(figsize=(10,6))
tables_data = df[df['Sub-Category'] == 'Tables']
plt.scatter(tables_data['Discount'], tables_data['Profit'], alpha=0.5, color='red')
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel('Discount %')
plt.ylabel('Profit ($)')
plt.title('Tables: Discount vs Profit (Red dots below zero = losses)')
plt.grid(True, alpha=0.3)
plt.show()
```
